```python
X = student_deatails[['study_hours',
                     'attendance_percentage',
                     'math_score',
                     'science_score',
                     'english_score']]

Y = student_deatails['overall_score']
X.head()
```

```python
Y.head()
```

```python
## The dataset is now prepared for splitting into training and testing sets.
```

```python
## Split the dataset into training and testing sets.
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
print(X_train.shape)

print(X_test.shape)

print(y_train.shape)

print(y_test.shape)
```

```python
## 80% of the dataset was used for training and 20% was used for testing.
## The model is now ready for training using unseen evaluation data.
```

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
print("Linear Regression Imported Successfully!")
```

```python
### The machine learning algorithm has been imported successfully and is ready to create the prediction model.
```

```python
model = LinearRegression()
student_model = LinearRegression()
print(model)
```

```python
## A Linear Regression model object was successfully created. At this stage, it has not learned from any data.
```

```python
model.fit(X_train, y_train)
```

```python
## vThe trained model is now ready to predict overall scores for unseen student data.
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

```

```python
y_pred = model.predict(X_test)
print(y_pred[:10])
```

```python
comparison = pd.DataFrame({
    "Actual Score": y_test,
    "Predicted Score": y_pred
})

comparison.head(10)
```

```python
## The predicted scores can now be compared with the actual scores to evaluate the model's performance.
```

```python
print(y_pred[:10])
```

```python
comparison = pd.DataFrame({
    "Actual Score": y_test,
    "Predicted Score": y_pred
})

comparison.head(10)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

```python
## A lower MAE indicates better prediction accuracy.
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
## A lower MSE indicates that the model predictions are closer to the actual values.
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_test, y_pred)
print("Root Mean Squared Error:", rmse)
```

```python
## A lower RMSE indicates that the model predicts student scores more accurately.
```

```python
from sklearn.metrics import mean_squared_error
import math

rmse = math.sqrt(mean_squared_error(y_test, y_pred))

print("Root Mean Squared Error:", rmse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```

```python
evaluation = pd.DataFrame({
    "Metric": ["MAE", "MSE", "RMSE", "R² Score"],
    "Value": [mae, mse, rmse, r2]
})

evaluation
```

```python
student.figure(figsize=(8,6))

student.scatter(y_test, y_pred, alpha=0.6)

student.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red',
         linewidth=2)

student.title("Actual Score vs Predicted Score", fontsize=15)
student.xlabel("Actual Score")
student.ylabel("Predicted Score")
student.grid(True)
student.show()
```
